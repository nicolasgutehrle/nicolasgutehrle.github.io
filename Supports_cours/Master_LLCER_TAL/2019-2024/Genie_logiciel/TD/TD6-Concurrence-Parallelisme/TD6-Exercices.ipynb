# Exercice 1

A l'aide de multiprocessing, appliquez la méthode "fibonacci" ci-dessous à une liste de valeur allant de 0 à 1000. 

```python
def fibonacci(n):
    # partie par défaut : on dicte la condition de sortie
    # à savoir que n est égal à 0 ou 1
    if n in (0, 1):
        return n 
    # partie récursive : on appelle la fonction au sein d'elle-même :
    # on va plus profond.
    else:
        return fibonacci(n-1) + fibonacci(n-2)
```
