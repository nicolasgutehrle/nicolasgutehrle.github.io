 Ajoutez le plus de commentaires possible ainsi que la documentation à votre classe **Voiture**. 
 

 A la fin du script, affichez la documentation de la classe ``Voiture``, puis exécutez le script.

# Solution exercice 5

```python
class Voiture:

    """
    Représentation d'une Voiture
    
    :param marque: la marque de la voiture
    :type marque: str
    :param moteur: le type de moteur (diesel, essence, électrique)
    :type moteur: str
    :param nbr_portes: le nombre de portes
    :type nbr_portes: int
    :param boite: le type de boite (auto, manuelle)
    :type boite: str
    """

    __nbr_roues = 4

    def __init__(self, marque, moteur, nbr_portes, boite):
        """
        Constructeur de la classe Voiture
        """
        self.__marque = marque
        self.__moteur = moteur
        # le nombre de portes va définir la valeur de l'attribut prix
        self.__nbr_portes = self.setNbrPortes(nbr_portes)
        self.__boite = boite
    
    def getMarque(self):
        """
        Getter de l'attribut marque
        """
        return self.__marque
    
    def getMoteur(self):
        """
        Getter de l'attribut moteur
        """
        return self.__moteur
    
    def getPortes(self):
        """
        Getter de l'attribut portes
        """
        return self.__nbr_portes

    def getBoite(self):
        """
        Getter de l'attribut boite
        """
        return self.__boite
    
    def getPrix(self):
        """
        Getter de l'attribut prix
        """
        return self.__prix
    
    def setMarque(self, marque):
        """
        Setter de l'attribut marque
        :param marque: la nouvelle marque
        :type marque: str 
        """
        self.__marque = marque
    
    def setMoteur(self, moteur):
        """
        Setter de l'attribut moteur
        :param moteur: le nouveaux type moteur
        :type moteur: str 
        """
        self.__moteur = moteur 

    def setNbrPortes(self, nvx_nbr_portes):
        """
        Setter de l'attribut NbrPortes. Change également la valeur de l'attribut prix
        :param nvx_nbr_portes: le nouveau nombre de portes
        :type nvx_nbr_portes: int 
        """
        self.__nbr_portes = nvx_nbr_portes
        
        # le nombre de porte définit le prix
        if self.__nbr_portes <= 3:
            self.setPrix(30000)

        elif self.__nbr_portes == 4:
            self.setPrix(40000)

        else:
            self.setPrix(50000)

    def setBoite(self, boite):
        """
        Setter de l'attribut boite.
        :param boite: nouveau type de boite 
        :type boite: str
        """
        self.__boite = boite

    def setPrix(self, nvxPrix):
        """
        Setter de l'attribut prix.
        :param boite: nouveau prix 
        :type boite: int
        """
        self.__prix = nvxPrix

    @staticmethod
    def klaxon():
        """
        Methode de classe qui affiche le mot "klaxon !"
        """
        print('Klaxon !')

    def afficher_details(self):
        """
        Affiche un dictionnaire avec les attributs et leurs valeurs
        :return: les attributs de l'objet
        :rtype: dict
        """
        proprietes = {
            "Marque": self.__marque,
            "Moteur": self.__moteur,
            "Nombre_portes": self.__nbr_portes,
            "Boite": self.__boite,
            "Prix": self.__prix,
            "Roues": Voiture.__nbr_roues
        }
        return proprietes


        
```
