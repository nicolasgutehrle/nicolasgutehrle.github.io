Importation des données :

```python
import pandas as pd

# NICOLAS: je n'avais pas les fichiers anglais de mon côté, donc pour la correction, je prend les mêmes que pour les cours
df_train = pd.read_csv('data/multiclass/as_train.csv')
df_dev = pd.read_csv('data/multiclass/as_dev.csv')
df_test = pd.read_csv('data/multiclass/as_test.csv')
# df_train = pd.read_csv('data/multiclass_en/as_train.csv')
# df_dev = pd.read_csv('data/multiclass_en/as_dev.csv')
# df_test = pd.read_csv('data/multiclass_en/as_test.csv')
```

Première étape : vectorisation des données textuelles

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# NICOLAS : Du coup, si tu traites les fichiers en anglais, ce n'est pas du fichier stopwords en français dont tu vas avoir besoin

with open('data/fr_stopwords.txt') as f:
    stopwords = [line.strip() for line in f]
    
data = df_train['texts']
```

```python
tfidf = TfidfVectorizer(stop_words=stopwords, max_features=10000)
X = tfidf.fit_transform(data)
X.shape
```

Clustering :

```python
from sklearn.cluster import MiniBatchKMeans

mbkm = MiniBatchKMeans(n_clusters=20) #on regroupe les données dans 20 clusters
                                        
centers = mbkm.fit_transform(X)
centroids = mbkm.cluster_centers_
```

```python
# on tente de predire a quel cluster un nouveau document appartient
first_doc = df_dev['texts'].iloc[0]
print("First doc :\n", first_doc)
vec_first_doc = tfidf.transform([first_doc])
pred = mbkm.predict(vec_first_doc)
print("First doc appartient au cluster :", pred[0])
```

On entraîne notre modèle avec Pipeline :

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from joblib import dump

# ici on se contente simplement de renommer les variables
# sans les faire passer par CountVectorizer
X_train, y_train = df_train['texts'], df_train['classes']
X_dev, y_dev = df_dev['texts'], df_dev['classes']

# NICOLAS: idem je suppose que tu n'as pas pu aller jusque là. Du coup tu as simplement copié collé la Pipeline, sans faire de modification
pipe = Pipeline(
    [
     ('vectorizer', TfidfVectorizer(stop_words=stopwords, max_features=10000)), # 1ere couche : vectorizer
     ('linear', MiniBatchKMeans(n_clusters=17)) # Derniere couche : modele
    ]
)
pipe.fit(X_train, y_train)

# pipe = Pipeline(
#     [
#      ('vectorizer', CountVectorizer(ngram_range=(2,2))), # 1ere couche : vectorizer
#      ('linear', LogisticRegression()) # Derniere couche : modele
#     ]
# )

# on entraine le Pipeline comme si l'on entrainait un modele
# score = pipe.score(X_dev, y_dev)
# print("Score :", score)

# NICOLAS : sans faire exprès, je lui ai fait calculer le score: regarde à quel point le modèle est mauvais en classification !
```

En utilisant le même dataset, vous utiliserez la librairie ``gensim`` pour entraîner un modèle de Topic Modelling. Vous utiliserez l'algorithme LDA, comme vu en cours. Une fois le modèle entraîné, vous afficherez les 10 premiers topics obtenus. Vous utiliserez ensuite votre modèle pour prédire le topic d'un document issu du devset ou du testset.

On commence par installer la librairie gensim :

```python
!pip3 install gensim
```

On fait le prétraitement du texte :

```python
with open('data/fr_stopwords.txt') as f:
    stopwords = [line.strip() for line in f]

def tokenize(data):
    """
    Simple tokenizer pour l'exemple
    """
    new_data = [text.lower() for text in data]
    new_data = [text.split() for text in new_data] # tokenize
    new_data = [[word for word in doc if word not in stopwords]
               for doc in new_data]
    return new_data
corpus = tokenize(df_train['texts'])
```

```python
from gensim.corpora import Dictionary

# Dictionary constitue notre vocabulaire, associe un id a chaque token
# ainsi que différentes statistiques sur les documents du corpus
dictionary = Dictionary(corpus)
```

```python
# on vectorize notre corpus selon la methode BoW (fréquence d'apparition d'un token)
vec_corpus = [dictionary.doc2bow(doc) for doc in corpus]
```

Maintenant nous pouvons entraîner notre modèle. En entrée de celui-ci nous avons le corpus vectorisé, le dictionnaire et le nombre de topics à identifier. On utilise l'algorithme LDA :

```python
from gensim.models.ldamodel import LdaModel
# NICOLAS: Non. Num_topics correspond au nombre de topic dans lesquels tu veux que l'algorithme regroupe les données. Ca ne correspond pas aux 10 premiers topics.
NUM_TOPICS = 10 #permet de trouver les 10 premiers topics
lda = LdaModel(corpus=vec_corpus, num_topics=NUM_TOPICS, id2word=dictionary)
```

Affichage des 10 premiers topics :

```python
lda.print_topics()
```

Prédiction du topic pour le document issu du devset :

```python
#Traitement du corpus
dev_texts = tokenize(df_dev['texts'])
vec_dev = [dictionary.doc2bow(text) for text in dev_texts]

## on predit le topic du premier document de dev
lda[vec_dev[0]]
```
