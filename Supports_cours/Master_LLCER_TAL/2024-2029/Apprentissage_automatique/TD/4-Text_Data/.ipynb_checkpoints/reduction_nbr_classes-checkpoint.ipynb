# Réduction du nombre de classes

En général, plus il y a de classes différentes et plus le problème est difficile à résoudre. 

Souvent les problèmes d'analyse de sentiments tentent de classer une donnée comme **positive** ou **négative** (ex: Imdb Dataset). Cependant, on trouve également des jeux de données (ex: Amazon Review) où le sentiment exprimé est plus subtil et repose sur 5 classes différentes : 

**Très mauvais**, **Mauvais**, **Neutre**, **Bon**, **Très bon** 

Suivant l'objectif et le niveau de finesse que l'on souhaite atteindre, on peut faire de l'analyse de sentiments **binaire** ou **multiclasses**

## Réduction à deux classes (positif ou négatif)

On considère tout texte ayant une note supérieur 3 comme positif, et tout texte ayant une note inférieure à 3 comme négatif. Tout texte noté 3 (donc neutre) ne sont pas conservés.

```
 filtered_data = data[data['label'] != 3] # on ne conserve que les données qui ne sont pas annotées "3"
 filtered_data.replace([1,2,4,5], [0, 0, 1, 1], inplace= True) # on remplace les valeurs 1 et 2 par 0 (négatif), 4 et 5 par 1 (positif). Inplace = True remplace les données directement
```

**Amazon Reviews** contient des jeux de données d'entraînement (train), de validation (val / dev) et de test (test). Tous les jeux de données ne proposent pas autant de fichier : en général même, seul le fichier d'entraînement est disponible. Dans ce cas là, il faut constituer les fichiers de validation et de test à partir du fichier d'entraînement.

Puisque l'on transforme le fichier d'entraînement pour n'avoir que deux classes, il ne faut pas oublier d'appliquer les mêmes transformations aux fichiers de validation et de tests. L'opération étant identique pour chaque fichier, on va écrire une fonction pour automatiser ce processus.

Notre fonction reduce_classes() doit pouvoir ouvrir n'importe quel fichier de 
données (idéalement csv ou json) avec pandas. On va donc créer une classe open_datafile() qui ouvre et retourne un de ces fichiers. La fonction retourne une erreur si le fichier n'est ni un json ou un csv.


## Ouverture du fichier

Nous allons écrire une fonction open_datafile() permettant de sélectionner un fichier sur le disque, de détecter son extension et de l'ouvrir si cette extension est autorisée. La fonction retournera un DataFrame pour manipuler le fichier avec **pandas**.

### Trouver l'extension du fichier

```
import os # os est une librairie permettant de manipuler le système d'exploitation de la même manière que sur un terminal.
filepath = 'amazon_reviews/train/dataset_fr_train.json'

# splitext() divise le nom du fichier de son extension, et retourne une liste contenant ces deux éléments
file_extension = os.path.splitext(filepath)[1] # l'extension du fichier est le second élément de la liste
print("File extension :", file_extension)
```

### Contrôler l'extension du fichier

```
# allowed_extension est un tuple contenant les extensions de fichier autorisées
# pour cette fonction. Comme splitext conserve le point dans le nom de fichier,
# les extensions dans allowed_extensions en ont un également.
allowed_extensions = ('.csv', '.json')

if file_extension in allowed_extensions: # vérifie que le type de fichier est autorisé
  if file_extension == ".json":
    df = pd.read_json(filepath, lines=True)
  else:
    df = pd.read_csv(filepath)
  
  print('Fichier chargé avec succès !')
  print(df.head())    
else:
  # le programme retourne une erreur et s'arrête automatiquement
  raise Exception("Le type de ce fichier n'est pas reconnu. Seuls les fichiers {} sont autorisés.".format(allowed_extensions))
```

### Fonction open_datafile()

```
import os
def open_datafile(filepath):
  """
  Ouvre un fichier avec pandas, en fonction de son extension
  """
  # splitext() divise le nom du fichier de son extension, et retourne une liste contenant ces deux éléments
  file_extension = os.path.splitext(filepath)[1] # l'extension du fichier est le second élément de la liste
  print("File extension :", file_extension)

  # allowed_extension est un tuple contenant les extensions de fichier autorisées
  # pour cette fonction. Comme splitext conserve le point dans le nom de fichier, les
  # extensions dans allowed_extensions en ont un également.
  allowed_extensions = ('.csv', '.json')

  if file_extension in allowed_extensions: # vérifie que le type de fichier est autorisé
    if file_extension == ".json":
      df = pd.read_json(filepath, lines=True)
    else:
      df = pd.read_csv(filepath)
    
    return df
    
  else:
    # le programme retourne une erreur et s'arrête automatiquement
    raise Exception("Le type de ce fichier n'est pas reconnu. Seuls les fichiers {} sont autorisés.".format(allowed_extensions))
```

## Réduire le nombre de classes

### Remplacer les valeurs existantes

```
filepath = 'amazon_reviews/train/dataset_fr_train.json'

df = open_datafile(filepath) # charge le fichier de données avec la fonction que l'on vient de créer
df.head()
```

```
replaced_classes = [1,2,4,5] # Noms des classes que l'on souhaite renommer
replacing_classes = [0, 0, 1, 1] # Valeurs par lesquelles on souhaite renommer les classes
df.replace(replaced_classes, replacing_classes, inplace= True) # inplace= True permet d'effectuer les modifications directement
df.head()
```

Si on regarde la colonne **stars**, on peut voir que tous les 1 ont été remplacés par 0

### Filtrer le DataFrame

On ne souhaite avoir que des données annotées positive ou négative. Puisque l'on ne peut pas réellement associée des données annotées comme neutres à l'une de ces deux classes, on fait le choix de ne pas les conserver
*texte en italique*

```
# Coup d'oeil rapide aux textes annotés comme **neutres** 
df[df['stars'] == 3].head()
```

```
ignored_classes = [3]
mask = df['stars'].isin(ignored_classes) # Serie de booléens indiquant pour chaque élément s'il est présent dans ignored_classes
print("Mask :\n", mask.head())

print('\n** ** ** **\n')

print("Aperçu de df avec le filtre mask :\n", df[mask].head() )
```

```
mask = ~mask # Puisque l'on veut toutes les valeurs qui n'appartiennent pas à 'ignored_classes', on inverse toutes les valeurs de mask
print("Mask inversé :", mask.head())

print('\n** ** ** **\n')

# on filtre df par rapport aux valeurs de 'mask'
print("Aperçu de df avec le filtre mask inversé:\n", df[mask].head() )
```

### Fonction reduce_classses()

```
def reduce_classes(filepath, replaced_classes, replacing_classes, ignored_classes = None, class_column = None):
  """
  Ouvre un fichier de données et le retourne en réduisant le nombre de classes dedans
  filepath: Nom du fichier à ouvrir
  replaced_classes: liste des classes à remplacer
  replacing_classes: liste des classes qui vont remplacer celles de 'replaced_classes'.
  Doivent être indiquées dans le même ordre que replaced_classes.
  ignore_classes: liste de classes que l'on ne souhaite ni conserver ni remplacer.
  class_column: nom de la colonne dans le fichier contenant les classes.
  """
  df = open_datafile(filepath) # charge le fichier de données avec la fonction que l'on vient de créer
  df.replace(replaced_classes, replacing_classes, inplace= True) # inplace= True permet d'effectuer les modifications directement
  if ignored_classes:
    mask = df[class_column].isin(ignored_classes) # Serie de booléens indiquant pour chaque élément s'il est présent dans ignored_classes
    mask = ~mask # Puisque l'on veut toutes les valeurs qui n'appartiennent pas à 'ignored_classes', on inverse toutes les valeurs de mask
    df = df[mask] # on filtre df par rapport aux valeurs de 'mask'
  
  return df
```

## Transformation des données

```
replaced_classes = [1,2,4,5] # Noms des classes que l'on souhaite renommer
replacing_classes = [0, 0, 1, 1] # Valeurs par lesquelles on souhaite renommer les classes
ignored_classes = [3] # Classes que l'on ne souhaite pas conserver dans le fichier final
class_column = 'stars' # Nom de la colonne du DF contenant les classes

# Ci-dessous, chemins vers les fichiers de données d'entraînement, de validation et de test
train_path = 'amazon_reviews/train/dataset_fr_train.json'
test_path = 'amazon_reviews/test/dataset_fr_test.json'
dev_path = 'amazon_reviews/dev/dataset_fr_dev.json'
```

```
binary_train = reduce_classes(train_path, replaced_classes, replacing_classes, ignored_classes, class_column)
binary_test = reduce_classes(test_path, replaced_classes, replacing_classes, ignored_classes, class_column)
binary_dev = reduce_classes(dev_path, replaced_classes, replacing_classes, ignored_classes, class_column)
```
