# Entraîner un modèle de classification binaire

## Chargements des données

```
# chemin vers les données d'entraînement, de validation et de test
train_path = '/content/drive/My Drive/Cours/Classification_binaire/data/binary_dataset_fr_train.json'
dev_path = '/content/drive/My Drive/Cours/Classification_binaire/data/binary_dataset_fr_dev.json'
test_path = '/content/drive/My Drive/Cours/Classification_binaire/data/binary_dataset_fr_test.json'
```

```
import pandas as pd
# on charge le fichier contenant les données d'entraînement
df_train = pd.read_csv(train_path)
df_dev = pd.read_csv(dev_path)
df_test = pd.read_csv(test_path)

df_train.head()
```

```
df_train = df_train[['review_body', 'stars']]
```

```
# Seules les colonnes 'review_body' et 'stars' nous intérèssent. On en profite
# pour renommer ces colonnes.
df_train = df_train[['review_body', 'stars']]
df_train.columns = ['texts', 'labels']

df_dev = df_dev[['review_body', 'stars']]
df_dev.columns = ['texts', 'labels']

df_test = df_test[['review_body', 'stars']]
df_test.columns = ['texts', 'labels']
```

```
fr_stopwords = []
with open('/content/drive/My Drive/Cours/Classification_binaire/data/fr_stopwords.txt') as f:
  for line in f:
    # on enleve les sauts de lignes et eventuels espaces en trop
    line = line.strip()
    fr_stopwords.append(line)
fr_stopwords[:10]
```

### Train model - Occurrences

```
train, labels = df_train['texts'], df_train['labels']
```

```
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(stop_words=fr_stopwords)
X = vectorizer.fit_transform(train)
```

```
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X, labels)
```

### Validation

```
from sklearn.metrics import accuracy_score, confusion_matrix

val, labels = vectorizer.transform(df_dev['texts']), df_dev['labels']
predictions = model.predict(val)
acc = accuracy_score(predictions, labels)
confmatrix = confusion_matrix(predictions, labels)
print(f"Accuracy : {acc}")
print("Confusion matrix :")
print(confmatrix)
```

```
import matplotlib.pyplot as plt
from sklearn.metrics import plot_confusion_matrix

class_names = [0, 1] # on indique les classes pour mieux comprendre la matrice
cmap = plt.cm.Blues # change les couleurs par défaut de la matrice

# plot_confusion_matrix prend le modele et les données de tests et fait une
# prediction sur ces données, pour ensuite afficher le résultat 
plot_confusion_matrix(model, val, labels,
                      display_labels = class_names,
                      cmap = cmap,
                      normalize = 'true' # permet de situer les résultats dans une échelle entre 0 et 1
                      )
```

### Sauvegarder un modèle

**pickle** et **joblib** sont deux librairies permettant de sauvegarder des données au format binaire sur le disques. On peut les utiliser pour sauvegarder un modèle et le charger plus tard (mais l'on peut également sauvegarder d'autres types de données.)

Historiquement, **joblib** est plus adapté pour sauvegarder de grosses masses de données, même si la différence n'est plus tout à fait vraie aujourd'hui. Cependant, sauvegarder un fichier avec **joblib** est syntaxiquement plus léger qu'avec **pickle**.

Voir la documentation de scikit-learn pour plus d'exemples : https://scikit-learn.org/stable/modules/model_persistence.html

```
from joblib import dump, load
dump(vectorizer, '/content/drive/My Drive/Cours/Classification_binaire/vectorizer.joblib')
dump(model, '/content/drive/My Drive/Cours/Classification_binaire/model.joblib')
```

```
import pickle
with open('/content/drive/My Drive/Cours/Classification_binaire/vectorizer.pickle', "wb") as f:
  pickle.dump(vectorizer, f)

with open('/content/drive/My Drive/Cours/Classification_binaire/model.pickle', "wb") as f:
  pickle.dump(model, f)
```

## Pipeline

(LEUR FAIRE SAUVEGARDER UN MODELE ET LE METTRE EN PRODUCTION AVANT DE LEUR PARLER DES PIPELINES)

L 'intérêt d'entraîner un modèle est de pouvoir l'utiliser en production pour prédire de nouveaux résultats sur des données qu'il n'a jamais vu jusque là. Que l'on traitre des données numériques ou textuelles, ces nouvelles données doivent passer par les mêmes étapes de pré-traitement (nettoyage, tokenization, ...) que les données d'entraînement. 

Ce processus étant systématiquement le même, il peut être laborieux de réécrire ces étapes à chaque fois que l'on utilise un modèle dans un projet. Pour simplifier l'utilisation d'un modèle en production, scikit-learn permet de définir une **pipeline** qui s'occupera d'appliquer ce processus aux nouvelles données.

Pipeline permet d'appliquer une série de transformation à des données avant de les envoyer vers un modèle pour prédictions. L'objet Pipeline peut être sauvegardé comme un autre modèle.

Documentation: https://scikit-learn.org/stable/modules/generated/sklearn.pipeline.Pipeline.html#sklearn.pipeline.Pipeline


```
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from joblib import dump

X_train, y_train = df_train['texts'], df_train['labels']
X_test, y_test = df_test['texts'], df_test['labels']

# Pipeline prend en argument une liste de tuple
# Le 1er element du tuple est le nom de l'objet, le 2nd est l'objet lui-meme
pipe = Pipeline(
    [
     ('vectorizer', CountVectorizer()), # 1ere couche : transformer
     ('linear', LogisticRegression()) # Derniere couche : modele
    ]
)

# on entraine le Pipeline comme si l'on entrainait un modele
pipe.fit(X_train, y_train)
predictions = pipe.predict(X_test)
"Precision :", accuracy_score(predictions, y_test)
dump(pipe, '/content/drive/My Drive/Cours/Classification_binaire/linear_pipeline.joblib')
```

```
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from joblib import dump

X_train, y_train = df_train['texts'], df_train['labels']
X_test, y_test = df_test['texts'], df_test['labels']

# Le resultat est le meme, mais il y a plus de fichiers à sauvegarder sur le
# disque et plus de risque d'oublier des étapes.
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(X_train)

linear = LogisticRegression()
linear.fit(X_train, y_train)

predictions = linear.predict(X_test)
"Precision :", accuracy_score(predictions, y_test)
dump(vectorizer, '/content/drive/My Drive/Cours/Classification_binaire/vectorizer.joblib')
dump(linear, '/content/drive/My Drive/Cours/Classification_binaire/linear_model.joblib')
```
