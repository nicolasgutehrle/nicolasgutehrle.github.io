
Notre objectif est de créer un modèle capable de classer un texte comme positif ou négatif. Pour entraîner un tel modèle, nous avons besoin de deux données: les textes eux-mêmes et leur classe (positif ou négatif)

** A DVLP **

```
import pandas as pd
# on charge le fichier json à l'aide de pandas
# le fichier json comporte plusieurs lignes : on specifie l'option 'lines=True', sinon on a une ValueError
amazon_data = pd.read_json('amazon_reviews/train/dataset_fr_train.json', lines=True)
amazon_data.head() # on regarde les 5 premières lignes du fichier
```

Ce fichier comporte plusieurs informations, notamment l'identifiant du produit, sa note, sa catégorie, etc... Cependant pour le projet de classification, seules deux informations nous intérèssent: **review_body** et **stars**. On ne va donc conserver que ces deux colonnes.




```
data = amazon_data[['review_body', 'product_category']] # on ne garde que ces deux colonnes
data.columns = ['text', 'label'] # on renomme les colonnes pour les manipuler plus facilement
data.head()
```

```
# il est toujours utile d'utiliser la fonction info() pour voir la quantité de données
# et surtout voir s'il y a autant de données dans chaque colonne
data.info()
```

## Occurences des mots

## Version Python

### Nested loops



```
from collections import Counter
counter = Counter() # Counter est un type de dictionnaire permettant de compter la frequence d'apparition d'un element
for text in data.text.values: # pour chaque texte contenu dans le dataset
  split_text = text.split() # on "tokenize" le texte en separant aux espaces
  for word in split_text: # pour chaque mot dans le texte, on compte + 1
    counter[word] += 1
print('Occurrence du mot "de" :', counter['de'])
```

### List comprehensions

Plus d'info sur les list comprehensions :

* Real Python : https://realpython.com/list-comprehension-python/
* Python for Beginners : https://www.pythonforbeginners.com/basics/list-comprehensions-in-python



```
from collections import Counter
counter = Counter([y for text in data.text.values for y in text.split()]) # idem, mais en une seule ligne de code
print('Occurrence du mot "de" :', counter['de'])
```

```
from pandas import Series

values = [x[1] for x in counter.items()]
index = [x[0] for x in counter.items()]
word_frequencies = Series(values, index = index) # on transforme Counter en Series pour le visualiser avec pandas
```

## Version pandas

```
# str : donne accès aux méthodes pour manipuler les chaînes de caractères
# split : divise la chaîne de caractère
# expand : transforme l'object Serie obtenu par spllit en DataFrame
# stack : transforme le DataFrame en Serie
# value_counts() : compte le nombre d'occurence de chaque élément
word_frequencies = data.text.str.split(expand=True).stack().value_counts()
```

## Visualisation des occurrences de mots

```
fig, axes = plt.subplots(nrows=2, ncols=2, figsize = (15, 15))
zipf = word_frequencies.plot(ax=axes[0, 0])
most_frequent = word_frequencies[:10].plot.bar(ax=axes[0, 1])
less_frequent = word_frequencies[::-1][:10].plot.bar(ax=axes[1, 0])
```

Sans surprise, les mots les plus fréquents sont les "mots-outils" (également dits "mots vides" ou "stopwords") comme "de", "le", "la", "et", ... Ce sont par conséquent les mots les moins importants pour la compréhension du corpus (cf **Loi de Zipf**)
